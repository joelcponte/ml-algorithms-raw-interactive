```python
from random import *

def train_perceptron(train, output, n):
    T = 400
    w0 = 0
    t = 0
    m = train.shape[0]
    x = train.loc[:,train.columns != output]
    y = train.loc[:,output]
    w = pd.Series([0,0], index= x.columns)
    while t<T:
        i = randint(0, m-1)
        x_i = x.loc[i,:]
        y_i = y[i]
        if y_i*(w0+x_i.dot(w)) <= 0:
            w0 += n*y_i
            w = w + y_i*n*x_i
        t+=1
        print(w)
    return(w, w0)
        
```

```python
w, w0 = train_perceptron(train, "class", 1)
w0
```
